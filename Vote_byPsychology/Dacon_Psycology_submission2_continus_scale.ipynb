```python

def preprocess(df): # 함수로 만들어놓으면 test셋에 적용할때 편하다.
    df_new = df.copy()
        
    def normalize(data):
        return (data - mean) / stdev
    
#     def encode_category(data):
#         drop_enc = preprocessing.OneHotEncoder(drop='first', categories='auto').fit(data)
#         return drop_enc.transform(data).toarray()

    for col in df.columns:

        if df[col].dtype == 'float64':
            print(col, 'float64 -> normalize')
            mean = df[col].mean()
            stdev = df[col].std()
            new_col = '_norm_{}'.format(col)
            df_new[new_col] = df[col].apply(normalize)
        elif df[col].dtype == 'object':
            print(col, 'object -> category')
            new_col = '_cat_{}'.format(col)
            df_new[new_col] = df[col].astype('category')
        elif df[col].dtype == 'int' and 'Q' in col and 'E' in col:
            print(col, 'int -> normalize')
            mean = df[col].mean()
            stdev = df[col].std()
            new_col = '_norm_{}'.format(col)
            df_new[new_col] = df[col].apply(normalize)
        elif df[col].dtype == 'int' and 'tp' in col:
            print(col, 'int -> normalize')
            mean = df[col].mean()
            stdev = df[col].std()
            new_col = '_norm_{}'.format(col)
            df_new[new_col] = df[col].apply(normalize)
        elif df[col].dtype == 'int' and 'wr_' in col:
            print(col, 'int -> binary')
            new_col = '_norm_{}'.format(col)
            df_new[new_col] = df[col].astype('bool').astype('int')
        elif df[col].dtype == 'int' and 'wf_' in col:
            print(col, 'int -> binary')
            new_col = '_norm_{}'.format(col)
            df_new[new_col] = df[col].astype('bool').astype('int')
        elif df[col].nunique() < df.shape[0] ** 0.5:
            print(col, 'int -> category')
            new_col = '_cat_{}'.format(col)
            df_new[new_col] = df[col].astype('category')
        else:
            print(col, '? -> ?')
    
    cat_cols = [x for x in df_new.columns.tolist() if x.startswith('_cat_')]
    print("CAT cols: {}".format(cat_cols))
    
    df_new_encoded = pd.concat([df_new] + [pd.get_dummies(df_new[col], prefix="_norm{}".format(col)) for col in cat_cols], axis=1)
    print(df_new_encoded.shape)
    
    df_new_encoded.drop(columns=[x for x in df_new.columns.tolist() if not x.startswith('_norm_')], inplace=True)
    print(df_new_encoded.columns.tolist())
    print(df_new_encoded.shape)
    return df_new_encoded

train_x_new = preprocess(train_x)
train_x_new.head()
```

## 모델 만들기

```python
import keras
from keras import Input, Model
from keras.layers import Dense
```

```python
def create_dense_layer_model(n_block, n_layer, n_dim_input):
    x = Input(shape=(n_dim_input,))
    out = x
    
    out = Dense(1000, activation='relu', name='Input_dense')(out)
    # Block
    for block_id in range(n_block):
        bypass = out
        # Layer
        for layer_id in range(n_layer):
            out = Dense(1000, activation='relu', name='{}_{}dense'.format(block_id, layer_id))(out)
        
        out = keras.layers.Add()([bypass, out])
        
    out = Dense(2, activation='softmax', name='final_dense')(out)
    model = Model(x, out)
    return model

clf = create_dense_layer_model(3, 3, n_dim_input=X_train.shape[1])
op = keras.optimizers.Adam(lr=0.001)  # Optimizer
loss = keras.losses.CategoricalCrossentropy()
clf.compile(
    optimizer=op,
    loss=loss,
    metrics=[keras.metrics.CategoricalCrossentropy(), keras.metrics.CategoricalAccuracy()],
)
clf.summary()

history = clf.fit(
    X_train, 
    y_train_binary, 
    batch_size=256,
    epochs=20,
    validation_data=(X_val, y_val_binary),
)    

history
```

```python
clf2 = create_dense_layer_model(1, 3, n_dim_input=X_train.shape[1])
op = keras.optimizers.Adam(lr=0.001)  # Optimizer
loss = keras.losses.CategoricalCrossentropy()
clf2.compile(
    optimizer=op,
    loss=loss,
    metrics=[keras.metrics.CategoricalCrossentropy(), keras.metrics.CategoricalAccuracy()],
)
clf2.summary()

history = clf2.fit(
    X_train, 
    y_train_binary, 
    batch_size=256,
    epochs=20,
    validation_data=(X_val, y_val_binary),
)    

history
```
